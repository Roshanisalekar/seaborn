```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("ROSH.csv")
df.head()
```

```python
sns.histplot(df["Revenue"])
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(df["Revenue"], bins=5)
plt.title("Revenue Distribution (5 Bins)")
plt.show()
```
