```python
 import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("ecommerce_sales.csv")

sns.boxplot(y=df["Revenue"])
plt.title("Revenue Distribution")
plt.grid()
plt.show()
```

```python
df
```

```python
sns.boxplot(x="Category", y="Revenue", data=df)
plt.title("Revenue by Category")
plt.grid()
plt.show()
```

```python
sns.boxplot(x="Customer_Type", y="Revenue", data=df, palette="Set2")
plt.title("Customer Type vs Revenue")
plt.legend()
plt.show()
```
