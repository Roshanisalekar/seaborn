```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ecommerce_sales.csv")

sns.countplot(x="Category", data=df)
plt.title("Price vs Revenue")
plt.show()
```

```python
df = pd.read_csv("ecommerce_sales.csv")

sns.countplot(x="Customer_Type", data=df)
plt.title("Price vs Revenue")
plt.show()
```

```python
sns.countplot(x="Category", data=df)
plt.title("Product Category Count")
plt.show()
```

```python
# Count Plot with Hue (IMPORTANT 🔥)

sns.countplot(x="Category", hue="Customer_Type", data=df)
plt.title("Category vs Customer Type")
plt.show()
```

```python
sns.countplot(x="Region", data=df)
plt.title("Region-wise Count")
plt.show()
```

```python
sns.countplot(x="Category", data=df, palette="Set2")

plt.show()
```
