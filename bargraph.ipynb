```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("ecommerce_sales.csv")
df.head()
```

```python
sns.barplot(x="Product", y="Revenue", data=df)
plt.title("Product Wise Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

```python
sns.barplot(
    x="Product",
    y="Revenue",
    hue="Category",
    data=df
)

plt.title("Product Revenue by Category")
plt.show()
```

```python
sns.barplot(
    x="Region",
    y="Revenue",
    data=df
)

plt.title("Region Wise Revenue")
plt.show()
```

```python
sns.barplot(
    x="Revenue",
    y="Product",
    data=df
)
plt.title("Revenue by Product")
plt.show()
```

```python
sns.set_style("whitegrid")

sns.barplot(
    x="Product",
    y="Revenue",
    hue="Region",
    data=df,
    palette="coolwarm"
)

plt.title("Sales Analysis Dashboard")
plt.show()
```
