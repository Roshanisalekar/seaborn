```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("ecommerce_sales.csv")
```

```python
sns.pairplot(
    df[["Price", "Quantity", "Revenue", "Rating"]]
)
plt.show()
```

```python
sns.pairplot(
    df[["Price", "Quantity", "Revenue", "Rating"]]
)
plt.show()
```

```python
sns.pairplot(
    df,
    hue="Category"
)
plt.show()
```

```python
sns.pairplot(
    df,
    hue="Category",
    markers=["o", "s"]
)
plt.show()
```

```python
sns.pairplot(
    df,
    hue="Category",
    diag_kind="kde"
)
plt.show()
```

```python
sns.pairplot(df, hue="Category")
plt.show()
```
