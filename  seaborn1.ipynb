```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Data from the sample dataset
data_products = {
'category': ['Electronics','Electronics','Electronics','Electronics','Electronics','Electronics','Electronics','Electronics','Electronics','Electronics'],
'sale_price': [150, 50, 15, 25, 75, 180, 65, 12, 30, 85],
'store_location': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B']
}
df_products = pd.DataFrame(data_products)
# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_products, x='category', y='sale_price', hue='store_location')
# Add titles and labels
plt.title('Distribution of Sale Price by Category and Store')
plt.xlabel('Product Category')
plt.ylabel('Sale Price ($)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Data from the sample dataset
data_monthly = {
'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
'sales': [22, 25, 30, 35, 42, 45, 50, 48, 43, 38, 32, 28],
'ad_spend': [1.5, 1.8, 2.0, 2.5, 3.0, 3.2, 3.5, 3.3, 3.1, 2.7, 2.4, 2.0]
}
df_monthly = pd.DataFrame(data_monthly)
# 1. Create the correlation matrix (data must be 2D)
# We need to select numeric columns only
corr_matrix = df_monthly[['sales', 'ad_spend']].corr()
# 2. Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
# Add titles
plt.title('Correlation Matrix of Sales and Ad Spend')
plt.show()
```
