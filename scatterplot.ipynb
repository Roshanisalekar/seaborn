```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("ecommerce_sales.csv")

sns.scatterplot(x="Price", y="Revenue", data=df)
plt.title("Price vs Revenue")
plt.show()
```

```python
df
```

```python
sns.scatterplot(
    x="Price",
    y="Revenue",
    hue="Category",
    data=df
)
plt.title("Price vs Revenue by Category")
plt.show()
```

```python
sns.scatterplot(
    x="Price",
    y="Revenue",
    size="Quantity",
    data=df,
    sizes=(50, 200)
)
plt.title("Price vs Revenue (Size = Quantity)")
plt.show()
```

```python
sns.scatterplot(
    x="Price",
    y="Revenue",
    hue="Category",
    style="Customer_Type",
    data=df
)
plt.title("Sales Analysis")
plt.show()
```
