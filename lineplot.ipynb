```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("ecommerce_sales.csv")
df.head()
```

```python
sns.lineplot(x="Date", y="Revenue", data=df)
plt.title("Revenue Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.lineplot(
    x="Date",
    y="Revenue",
    data=df,
    marker="o"
)

plt.title("Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.lineplot(
    x="Date",
    y="Revenue",
    hue="Category",
    data=df,
    marker="o"
)

plt.title("Revenue by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.lineplot(
    x="Date",
    y="Revenue",
    hue="Region",
    data=df,
    marker="o"
)

plt.title("Region Wise Revenue Trend")
plt.xticks(rotation=45)
plt.show()
```
