```python
# 🔹 Without Hue

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("ecommerce_sales.csv")
sns.scatterplot(x="Price", y="Revenue", data=df)

plt.title("Revenue Distribution")
plt.grid()
plt.show()
```

```python
# 🔹 With Hue (Important)

sns.scatterplot(
    x="Price",
    y="Revenue",
    hue="Category",
    data=df
)
```

```python
#🔹 Hue with Multiple Categories

sns.scatterplot(
    x="Price",
    y="Revenue",
    hue="Customer_Type",
    data=df
)
```

```python
# 🔹 Hue + Style (Advanced)
sns.scatterplot(
    x="Price",
    y="Revenue",
    hue="Category",
    style="Customer_Type",
    data=df
)

```

```python
# 🔹 Hue with Bar Plot
sns.barplot(
    x="Category",
    y="Revenue",
    hue="Customer_Type",
    data=df
)
```
